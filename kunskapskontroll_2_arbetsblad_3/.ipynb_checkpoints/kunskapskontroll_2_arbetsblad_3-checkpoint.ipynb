````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
x = np.random.rand(100)
y1 = x + 0.1 * np.random.randn(100)  # Positive correlation
y2 = -x + 0.1 * np.random.randn(100)  # Negative correlation
y3 = 0.1 * np.random.randn(100)       # No correlation

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].scatter(x, y1)
axes[0].set_title('Figure 1')
axes[1].scatter(x, y2)
axes[1].set_title('Figure 2')
axes[2].scatter(x, y3)
axes[2].set_title('Figure 3')
````
